```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math, random
```

```python
df= pd.read_csv('Ads_CTR_Optimisation.csv')
```

```python
N = df.shape[0]
number_of_ads = df.shape[1]
ads_selected = []
number_of_rewards_0 = [0]*number_of_ads
number_of_rewards_1 = [0]*number_of_ads
total_reward =0

for i in range(N):
  ad =0
  max_random = 0
  for j in range(number_of_ads):
    random_beta_reward = random.betavariate(number_of_rewards_1[j] +1, + number_of_rewards_0[j] +1)
    if random_beta_reward > max_random:
      max_random = random_beta_reward
      ad = j
    
  ads_selected.append(ad)
  reward = df.values[i, ad]
  if reward ==0:
    number_of_rewards_0[ad] +=1

  else:
    number_of_rewards_1[ad] +=1
  
  total_reward += reward
```

```python
plt.hist(ads_selected)
plt.title('Histogram of Ads Selected')
plt.xlabel('Ads')
plt.ylabel('Number of times each ad was selected')
plt.show()
```
